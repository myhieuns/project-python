```python
data['Spent'] = data['MntWines'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntGoldProds'] + data['MntSweetProducts']
```

```python
data['Living_With'] = data['Marital_Status'].replace(
{
    'Married' : 'Partner',
    'Together': 'Partner',
    'Single' : 'Alone',
    'Divorced': 'Alone',
    'Widow' : 'Alone',
    'Alone': 'Alone'
})
```

```python
data['Children'] = data['Kidhome'] + data['Teenhome']
```

```python
data['Family_size'] = data['Living_With'].replace({"Alone":1, "Partner": 2}) + data["Children"]
```

```python
data["Education"]=data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

```python
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
s = (data1.dtypes == 'object')
```

```python
object_cols = list(s[s].index)
```

```python
object_cols
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
for i in object_cols:
    data1[i] = data1[[i]].apply(LE.fit_transform)
```

```python
data1.Education.value_counts()
```

```python
ds = data1.copy()
```

```python
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis = 1)
```

```python
ds
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds), columns = ds.columns)
```

```python
scaled_ds
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.fit_transform(scaled_ds), columns = (['col1', 'col2', 'col3']))
```

```python
PCA_ds.describe().T
```

```python
x = PCA_ds['col1']
y = PCA_ds['col2']
z = PCA_ds['col3']

fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111, projection = "3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
Elbow_M = KElbowVisualizer(KMeans(),  k = 10)
Elbow_M.fit(PCA_ds)
```

```python
from sklearn.cluster import AgglomerativeClustering
AC = AgglomerativeClustering(n_clusters = 4)
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds['Clusters'] = yhat_AC
data1['Cluster'] = yhat_AC
```

```python
fig = plt.figure(figsize = (10,8))
ax = plt.subplot(111, projection = "3d", label = 'bla')
ax.scatter(x,y,z, s = 40, c = PCA_ds['Clusters'], marker = "o", cmap = cmap)
```

```python
pal = ["#682F2F","#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=data1["Cluster"], palette= pal)
pl.set_title("Distribution Of The Clusters")
plt.show()
```

```python
pl = sns.scatterplot(data = data1,x=data1["Spent"], y=data1["Income"],hue=data1["Cluster"], palette= pal)
pl.set_title("Cluster's Profile Based On Income And Spending")
plt.legend()
plt.show()
```
